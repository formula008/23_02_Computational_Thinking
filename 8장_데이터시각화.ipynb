### range() 사용하기

그래프 겹쳐 그리고
- 꺾은선 그래프 겹쳐 그리기

````
import matplotlib.pyplot as plt

singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

plt.plot(singer, week1)
plt.plot(singer, week2)
plt.show()
````

````
plt.rc('font',family='NanumGothic')
plt.title('오디션 프로그램 득표 현황')
plt.plot(singer, week1, label = '첫째주', color='hotpink')
plt.plot(singer, week2, label = '둘째주', color='royalblue')
plt.legend()
plt.show()
````

- 막대그래프 겹쳐그리기(항아리모양)

````
for i in range(len(week1)) :
    week1[i] = -week1[i]

plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 표시
plt.rc('font', family='NanumGothic')
plt.title('오디션 프로그램 득표 현황')
plt.barh(singer, week1, label = '첫째주', color = 'hotpink')
plt.barh(singer, week2, label = '둘째주', color = 'royalblue')
plt.legend()
plt.show()
````

# 그래프 명령어 정리
**그래프 그리는 포맷**
```
import matplotlib.pyplot as plt

plt.title('그래프 제목')
# --- 그래프 형태에 따라 알맞은 명령어 넣기 ---
plt.show()
```
    

**그래프 유형별 명령어**
```
# ---- 그래프 유형별 명령어
# 꺾은선그래프
plt.plot([x축 리스트], [y축 리스트])

# 막대그래프 세로
plt.bar([x축 리스트], [y축 리스트])

# 막대그래프 가로
plt.barh([x축 리스트], [y축 리스트])

# 히스토그램 (x축, y축을 구분하지 않고 하나의 리스트만 넣음)
plt.hist( [리스트], bins = 구간개수 )

# 원 그래프 (x축, y축을 구분하지 않고 하나의 리스트만 넣음)
plt.pie( [리스트], labels = [데이터 구간 이름])

# 산점도
plt.scatter([x축 리스트], [y축 리스트])
```

**그래프 명령어 내에 추가 가능한 옵션**

```
color = '색상이름'
alpha = 투명도

# plt.plot(), plt.bar(), plt.barh(), plt.scatter()는 label 사용
label = '범례내용'

# plt.pie(), plt.boxplot()은 labels 사용
labels = [범례 리스트]

# plt.pie()에 퍼센트 자동 표시
autopct = ‘%1.1%%’
```

**기타 사용가능한 명령어**


```
plt.legend()            # 범례 나타내기
plt.xlabel()            # x축 이름 붙이기
plt.ylabel()            # y축 이름 붙이기
```
